```python
import os

import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.segmentation import mark_boundaries, slic
```

## Hare

```python
hare_bgr = cv2.imread('Zajac_kolor.jpg')[..., ::-1]
hare_hsv = cv2.cvtColor(hare_bgr, cv2.COLOR_RGB2HSV)
```

```python
def get_bouder(image, channel_axis, n_segments, sigma, hsv=False, image_show=False):
    image_slic = slic(image=image, channel_axis=channel_axis, n_segments=n_segments, sigma=sigma)
    if hsv:
        image = cv2.cvtColor(image, cv2.COLOR_HSV2RGB)
    if image_show is not False:
        new_channel = np.zeros_like(image[:, :, 0])

        # Połącz nowy kanał z istniejącymi kanałami
        new_image = cv2.merge([new_channel, image[:, :, 0], image[:, :, 1]])

        return mark_boundaries(new_image, image_slic, outline_color=(1,0,0))
    return mark_boundaries(image, image_slic, outline_color=(1,0,0))
        
```

# Group superpixels


```python
image = cv2.imread('Zajac_kolor.jpg')[..., ::-1]
n_segments = 50
sigma = 10
plt.figure(figsize=(100, 100))
image = image

def get_item(image, n_segments, sigma, channels=[0,1,2]):
    # slic na obrazie RGB
    image_slic = slic(image=image, n_segments=n_segments, sigma=sigma)
    unique_values = np.unique(image_slic)

    # wybranie kanałów HSV
    image_hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV).copy()
    selected_image = image_hsv[:, :, channels]
    
    # Uśrednianie wartośći pikseli dla każdego segmentu ze slic
    for segment in unique_values:
        mask = (image_slic == segment)
        pixels = selected_image[mask]
        means = np.mean(pixels, axis=0)
        selected_image[mask] = means

    # do float
    selected_image_float = np.float32(selected_image)
    
    # jeśli kanał H podziel przez 179 wpp dziel 255
    num_channels = len(channels) - 1
    if channels[0] == 0:
        selected_image_float[:, :, 0] = selected_image_float[:, :, 0] / 179
    else:
        selected_image_float[:, :, 0] = selected_image_float[:, :, 0] / 255
    num_channels = len(channels)
    for other_channels in range(1, num_channels):
        selected_image_float[:, :, other_channels] = selected_image_float[:, :, other_channels] / 255
        
    # oblicz KNN dla (lista pikseli, wartość piksla)
    Z = selected_image_float.reshape((-1,len(channels)))
    ret,label,center = cv2.kmeans(data=Z,
                                    K=2,
                                    bestLabels=None,
                                    criteria=(cv2.TERM_CRITERIA_EPS, 10, 1.0),
                                    attempts=10,
                                    flags=cv2.KMEANS_RANDOM_CENTERS)
    # segmentacja do rozmiaru zdjęcia
    segmentation = label.reshape((h, w))

    # odtwórz obraz HSV tzn wstaw wszedzie 255 jesli kanal był uzywany do usredniania piksli wstaw jego wartosc (tak wiem H = [0,179], a więc H =76)
    complete_image = np.ones_like(image_hsv) * 255
    used_channels = 0
    for x in range(3):
        if x in channels:
            complete_image[:, :, x] = selected_image[:, :, used_channels]
            used_channels += 1
    superpixel_image = cv2.cvtColor(complete_image, cv2.COLOR_HSV2RGB)
    
    
    boundries = mark_boundaries(image, segmentation, outline_color=(1,0,0))

    return ret, label, center, superpixel_image, segmentation, boundries
```

```python
from itertools import combinations

elements = [0, 1, 2]

subsets = []

# Itertools.product dla 1, 2, i 3-elementowych podzbiorów
for r in range(1, 4):
    subsets.extend(list(combinations(elements, r)))
```

```python
all_channels = [[]]
plt.figure(figsize=(200, 200))
rows = len(subsets)
for idx, subset in enumerate(subsets[:],start=0):
    ret, label, center, superpixel_image, segmentation, boundries   = get_item(image, n_segments, sigma, subset)
    print(subset)
    plt.subplot(rows, 3, idx*3+1)
    plt.imshow(superpixel_image)
    plt.title(f'Channels={subset}')
    
    plt.subplot(rows, 3, idx*3+2)
    plt.imshow(segmentation)
    plt.title(f'Channels={subset}')
    
    plt.subplot(rows, 3, idx*3+3)
    plt.imshow(boundries)
    plt.title(f'Channels={subset}')

plt.tight_layout()
plt.show()
```
